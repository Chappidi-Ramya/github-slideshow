```python
x=np.linspace(0,10,100)
plt.figure(1)

#subplot1
plt.subplot(2,2,1)
plt.title('linear')
plt.plot(x,x)

#subplot2
plt.subplot(2,2,2)
plt.title('cubic')
plt.plot(x,x**3)

plt.figure(2)
#subplot4
plt.subplot(2,2,1)
plt.title('log')
plt.plot(x,np.log(x))

#subplot5
plt.subplot(2,2,2)
plt.title('exponential')
plt.plot(x,x**2)
plt.show()
```
